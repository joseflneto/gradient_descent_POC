# This code is a simple implementation of the gradient descent algorithm using only numpy, we define a cost MSE as the cost function 

# Here, we are finding the slope and the intercept of a X, y

### In the gradient descent algorithm, we want minimize the loss function output using the input values dot the theta values (example NN weights or a linear regression slope), we plot a function of errors and iterate over this function until find the closest to 0 value of the derivative of this error function, that means the optimal value where if I keep updating the theta, the error will increase. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_descent(X, y, theta, alpha, num_iters):
    m = len(y)
    J_history = np.zeros((num_iters, 1))
    
    for i in range(num_iters):
        # Compute hypothesis (y = Ax + B)
        h = np.dot(X, theta)
        
        # Compute error
        error = h - y
        
        # Compute gradient (this formula is the derivative of the mse function)
        gradient = np.dot(X.transpose(), error) / m
        
        # Update theta
        theta = theta - alpha * gradient
        
        # Save the cost J in every iteration
        J_history[i] = mse(X, y, theta)
    
    return theta, J_history

def mse(X, y, theta):
    m = len(y)
    # (y = Ax + B)
    h = X*theta[1] + theta[0]
    J = np.sum((h - y) ** 2) / (2 * m)
    return J
```

```python
# Generate some sample data
X = 2 * np.random.rand(1000, 1)
y = 4 + 3 * X + np.random.randn(1000, 1)

# Add a column of ones to X to account for the intercept term
X_b = np.c_[np.ones((1000, 1)), X]

# Define the learning rate and number of iterations
alpha = 0.1
n_iterations = 50

# Initialize theta to random values
theta = np.random.randn(2, 1)

# Call the gradient descent function to find the best values of theta
theta, cost_history = gradient_descent(X_b, y, theta, alpha, n_iterations)

# Plot the cost function over iterations
plt.plot(range(n_iterations), cost_history)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.show()

# Plot the data points and the best-fit line
plt.scatter(X, y)
plt.plot(X, X_b.dot(theta), color='red')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```
